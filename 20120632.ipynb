```python
import pandas as pd
```

```python
df = pd.read_csv("/content/FastFoodNutritionMenuV3.csv")
df.head()
```

```python
# Kiểu dữ liệu mỗi cột
df.info()
```

```python
columns = df.columns
numeric_columns = columns[2:]

for col in numeric_columns:
  df[col] = pd.to_numeric(df[col], errors='coerce')
df.info()
```

```python
# Phân bố dữ liệu
# Giá trị tối đa, tối thiểu, độ lệch chuẩn
df.describe()
# --> Độ lệch chuẩn cao ==> Phân tán dữ liệu lớn
```

```python
# Tính toán phần trăm giá trị còn thiếu cho mỗi cột
missing_values_percentage = df.isnull().mean() * 100

missing_values_percentage
```
